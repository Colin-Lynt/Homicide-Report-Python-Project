### Data Wrangling

```python
df.shape
```

```python
# Dropping columns from the df
df_2 = df.drop(columns = ['ID', 'Ori', 'Source', 'StateName', 'FileDate', 'MSA', 'Incident'])
```

```python
# Checking the columns in df_2
df_2.head()
```

```python
# Renaming column headers in df_2
df_2.rename(columns = {'CNTYFIPS' : 'City'}, inplace = True)
```

```python
# Print df_2 after changes
df_2.head()
```

### Data Types

```python
# Checking data types if df_2
df_2.describe()
```

```python
# Checking data types
df_2.dtypes
```

```python
# Fixing error message & changing data type of year to int64
df_2['Year'] = df_2['Year'].astype('int64')
```

```python
# Cecking for anymore mixed types
for col in df_2.columns.tolist():
    weird = (df_2[[col]].applymap(type) != df_2[[col]].iloc[0].apply(type)).any(axis = 1)
    if len (df_2[weird]) > 0:
        print(col)
```

```python
# Checking the value counts to make sure the column is a Object
df_2['Subcircum'].value_counts(dropna = False)
```

```python
# Changing data type for Subcircum
df_2['Subcircum'] = df_2['Subcircum'].astype('object')
```

### Missing Values

```python
# Checking for missing values
df_2.isnull().sum()
```

```python
# Dropping Subcircum from the df_2
df_2 = df_2.drop(columns = ['Subcircum'])
```

```python
# Checking df after droped column
df_2.head()
```

### Duplicate Data

```python
# Checking for duplicate values in the df
df_dups = df_2[df_2.duplicated()]
df_dups
```

```python
# dropping duplicate values in the df 
df_2_no_dups = df_2.drop_duplicates()
df_2_no_dups.shape
```

### Outliers

```python
# Checking for outliers
df_2_no_dups.describe()
```

```python
# Checking outliers from victim age column
df_2_no_dups.loc[df_2_no_dups['VicAge'] > 150]
```

```python
# Fixing outliers from victim age column
df_2_no_dups.loc[df_2_no_dups['VicAge'] > 150, 'VicAge'] = np.nan
df_2_no_dups['VicAge'].max()
```

```python
# Checking outliers from Offenders age column
df_2_no_dups.loc[df_2_no_dups['OffAge'] > 150]
```

```python
# Fixing outliers from offenders age column
df_2_no_dups.loc[df_2_no_dups['OffAge'] > 150, 'OffAge'] = np.nan
df_2_no_dups['OffAge'].max()
```

```python
# Checking outliers from Offenders age column
df_2_no_dups.loc[df_2_no_dups['OffAge'] < 2]
```

```python
df_2_no_dups.drop(df_2_no_dups.loc[df_2_no_dups['OffAge'] <= 2].index, inplace = True)
```

```python
# Checking for anymore outliers
df_2_no_dups.describe()
```

```python
# Checking for missing values
df_2_no_dups.isnull().sum()
```

```python
# Setting the mean value for victim age
VicAge_mean_value = df_2_no_dups['VicAge'].mean()
```

```python
# Applying mean value to missing values
df_2_no_dups['VicAge'].fillna(VicAge_mean_value, inplace = True)
```

```python
# Setting the mean value for Offender age
OffAge_mean_value = df_2_no_dups['OffAge'].mean()
```

```python
# Applying mean value to missing values
df_2_no_dups['OffAge'].fillna(OffAge_mean_value, inplace = True)
```

```python
# Checking for missing values
df_2_no_dups.isnull().sum()
```
